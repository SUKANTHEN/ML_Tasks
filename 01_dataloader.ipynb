```python
"""
dataloader.ipynb
----------------
"""
import pickle

with open('db/email_campaigns.pkl', 'rb') as file:
    loaded_dataset = pickle.load(file)

print(loaded_dataset)
```

```python
type(loaded_dataset)
```

```python
import pandas as pd
# Initialize an empty list to store data
flat_data = []

# Iterate through the data and extract values
for type_dict in loaded_dataset:
    for type_name, examples in type_dict.items():
        if isinstance(examples, list):
            for example_list in examples:
                if isinstance(example_list, list):
                    for example in example_list:
                        row_data = {"Type": type_name}
                        row_data.update(example)
                        flat_data.append(row_data)
                elif isinstance(example_list, dict):
                    row_data = {"Type": type_name}
                    row_data.update(example_list)
                    flat_data.append(row_data)
        elif isinstance(examples, dict):
            row_data = {"Type": type_name}
            row_data.update(examples)
            flat_data.append(row_data)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(flat_data)

# Display the DataFrame
print(df)
```

```python
df.to_excel("db/dataset_v1.xlsx",index=False)
```
